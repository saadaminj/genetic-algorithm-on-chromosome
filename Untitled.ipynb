```python
import random
class genetic_algorithm:
    def __init__ (self):
        None
    def randomize(self,population=4,length=8,generations=2):
        step=0                                      #to keep track of steps 
        self.population=population                  #how many chromosomes to work on at a time
        self.length=length                          #length of each chromosomes i.e how many bits
        self.generations=generations                #after how many iterations should the chromosomes mutate/ change
        self.chromosome=[]                          #actual chromosomes to work on
        
        
        ##############CREATING INITIAL POPULATION##############
        
        for i in range(self.population):            #create list of chromosomes with the size specified in self.population
            self.chromosome.append("") 
            for j in range(self.length):        #fill each chromosome till self.length
                self.chromosome[i]=self.chromosome[i]+str(random.randint(0,1))  #concate chromosome string with random bits 0 or 1
            #print(self.fitness(self.chromosome[i]))
        while(True):
            step=step+1                          #increment the steps
            
            #print("Initial State")
            #print(self.chromosome)
            
            ##############   EVALUATION & SELECTION   ##############
            
            # first check if there is any chromosome that satisfies the goal condition that is ALL 1s
            
            for i in range(self.population):   ## EVALUATION ## if 
                if(self.fitness(self.chromosome[i])==(2**(self.length))-1):
                    #print(self.chromosome)
                    #print("Goal Achieved  in "+str(step)+" Steps")
                    return step               ## return number of steps if goal is found
            
            
            #remove the chromosome with the lowest fitness value      ## self.fitness for selection of the 
                                                                      ## chromosome with least fitness value
            self.chromosome.remove((sorted(self.chromosome,key=lambda x:self.fitness(x) ))[0])
            
            #fill out the place of missing chromosome (that is removed above) by randomly choosing from remaining chromosome population
            self.chromosome.append(self.chromosome[random.randint(0,len(self.chromosome)-1)])
            #print("After Removing Smallest and Replicating the random chromosome sequence")
            #print(self.chromosome)
            
            ##############   CROSSOVER   ##############
        
            c_length=int(self.length/2)   ## crossover chromosome by half of their length means 
            j=0                           ## 1st chromosome get 2nd half of 2nd chromosome while 
                                          ## 2nd chromosome gets 2nd half of 1st chromosome
            for i in range(int(self.population/2)):  ## crossover between every 1st and 2nd chromosome. 
                if(j+1==self.population):        
                    break
                temp=self.chromosome[j]
                self.chromosome[j] = self.chromosome[j][:c_length] + self.chromosome[(j+1)][c_length:]
                self.chromosome[j+1]=self.chromosome[j+1][:c_length] + temp[c_length:]
                j=j+2
            #print("After Crossover")
            #print(self.chromosome)
            
            
            ############## MUTATION #############
            
            #mutation after every 2nd iteration
            if(step%self.generations==0):       # if the step count passed after last mutation is equal to the 
                                                # self.generations(generation after n number of steps)
                for i in range(self.population):
                    for j in range(self.length):
                        if(self.chromosome[i][j]=='0'): ## change/ mutate the first 0 that appears in the chromosome in direction left to right
                            self.chromosome[i]=self.chromosome[i][:j]+'1'+(self.chromosome[i][j+1:] if j+1<self.length else "")
                            break
                #print("After Mutation")
                #print(self.chromosome)
                
            #########   SURVIVOR SELECTION   ########
            
            #Those chromosomes which survives these steps will move to next step
            
        
    def fitness(self,chromosome):
        int_value=0
        j=len(chromosome)-1
        for i in range(len(chromosome)):
            int_value=int_value+int(chromosome[j])*(2**i)
            j=j-1
        return int_value
x=genetic_algorithm()
epoch=100

pop=[]
result=[]
for w in range(4,11):
    sum1=0
    for i in range(epoch):
        sum1=sum1+x.randomize(w,8,1)
    avg=int(sum1/epoch)
    pop.append(w)
    result.append(avg)
    #print("Took "+str(avg)+" Steps on Average, With population = "+str(w)+", length = "+str(y)+", generations = "+str(z))


import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(pop, result)

ax.set(xlabel='Chromosome Population', ylabel='Average Steps',
       title='Performance as a function of Population Size with Generation = 1')
ax.grid()
#ax.set_xticks(np.arange(min(population), max(population)+1, 1))
plt.show()



pop=[]
result=[]
for w in range(4,11):
    sum1=0
    for i in range(epoch):
        sum1=sum1+x.randomize(w,8,2)
    avg=int(sum1/epoch)
    pop.append(w)
    result.append(avg)
    #print("Took "+str(avg)+" Steps on Average, With population = "+str(w)+", length = "+str(y)+", generations = "+str(z))


import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(pop, result)

ax.set(xlabel='Chromosome Population', ylabel='Average Steps',
       title='Performance as a function of Population Size with Generation = 2')
ax.grid()
#ax.set_xticks(np.arange(min(population), max(population)+1, 1))
plt.show()




pop=[]
result=[]
for w in range(4,11):
    sum1=0
    for i in range(epoch):
        sum1=sum1+x.randomize(w,8,3)
    avg=int(sum1/epoch)
    pop.append(w)
    result.append(avg)
    #print("Took "+str(avg)+" Steps on Average, With population = "+str(w)+", length = "+str(y)+", generations = "+str(z))


import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(pop, result)

ax.set(xlabel='Chromosome Population', ylabel='Average Steps',
       title='Performance as a function of Population Size with Generation = 3')
ax.grid()
#ax.set_xticks(np.arange(min(population), max(population)+1, 1))
plt.show()









pop=[]
result=[]
for w in range(8,13):
    sum1=0
    for i in range(epoch):
        sum1=sum1+x.randomize(8,w,2)
    avg=int(sum1/epoch)
    pop.append(w)
    result.append(avg)
    #print("Took "+str(avg)+" Steps on Average, With population = "+str(w)+", length = "+str(y)+", generations = "+str(z))


import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(pop, result)

ax.set(xlabel='Chromosome Length', ylabel='Average Steps',
       title='Performance as a function of Length with Generation = 2')
ax.grid()
#ax.set_xticks(np.arange(min(population), max(population)+1, 1))
plt.show()


pop=[]
result=[]
for w in range(8,13):
    sum1=0
    for i in range(epoch):
        sum1=sum1+x.randomize(8,w,3)
    avg=int(sum1/epoch)
    pop.append(w)
    result.append(avg)
    #print("Took "+str(avg)+" Steps on Average, With population = "+str(w)+", length = "+str(y)+", generations = "+str(z))


import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(pop, result)

ax.set(xlabel='Chromosome Length', ylabel='Average Steps',
       title='Performance as a function of Length with Generation = 3')
ax.grid()
#ax.set_xticks(np.arange(min(population), max(population)+1, 1))
plt.show()
```
